```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/creditcard.csv')

# Quick overview of the data
print("Data shape:", data.shape)
print(data.head())

# Separate features and target variable. The 'Class' column is the label (0: non-fraud, 1: fraud)
X = data.drop('Class', axis=1)
y = data['Class']

# Scale the 'Amount' and 'Time' features
scaler = StandardScaler()
X['Amount_scaled'] = scaler.fit_transform(X[['Amount']])
X['Time_scaled'] = scaler.fit_transform(X[['Time']])

# Drop the original 'Amount' and 'Time' columns
X = X.drop(['Amount', 'Time'], axis=1)

# Split the data into training and testing sets (stratify by y to preserve class distribution)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)

# Initialize the Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# Train the model
rf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf.predict(X_test)
y_pred_proba = rf.predict_proba(X_test)[:, 1]

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

roc_auc = roc_auc_score(y_test, y_pred_proba)
print("ROC AUC Score:", roc_auc)

# Calculate and print accuracy in percentage
accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy: {:.2f}%".format(accuracy))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/creditcard.csv')

# Quick overview of the data
print("Data shape:", data.shape)
print(data.head())

# Separate features and target variable. The 'Class' column is the label (0: non-fraud, 1: fraud)
X = data.drop('Class', axis=1)
y = data['Class']

# Scale the 'Amount' and 'Time' features
scaler = StandardScaler()
X['Amount_scaled'] = scaler.fit_transform(X[['Amount']])
X['Time_scaled'] = scaler.fit_transform(X[['Time']])

# Drop the original 'Amount' and 'Time' columns
X = X.drop(['Amount', 'Time'], axis=1)

# Split the data into training and testing sets (stratify by y to preserve class distribution)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)

from sklearn.svm import SVC

svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
y_pred_proba = svm.predict_proba(X_test)[:, 1]

# Evaluate the model
print("SVM Report:")
print(classification_report(y_test, y_pred))

roc_auc = roc_auc_score(y_test, y_pred_proba)
print("ROC AUC Score:", roc_auc)

# Calculate and print accuracy in percentage
accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy: {:.2f}%".format(accuracy))
```
